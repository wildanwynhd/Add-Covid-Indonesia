<h4 style='color:Yellow'> Convert Data Types </h4>

```python
df.dtypes
```

```python
import datetime as dt

# covert data types 'key_as_string' from object to datetime
#df['key_as_string'] = pd.to_datetime(df['key_as_string'], format='%Y-%m-%d')

# change format 'key_as_string'
df['key_as_string'] = pd.to_datetime(df['key_as_string']).dt.date

# Normalize 'key_as_string'
df['key_as_string'] = pd.to_datetime(df['key_as_string']).dt.normalize()


# check data types 'key_as_string'
df.dtypes['key_as_string']
```
