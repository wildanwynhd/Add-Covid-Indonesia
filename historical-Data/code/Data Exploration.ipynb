```python
%store -r df
%store -r df_month
%store -r df_week

data = df
data_month = df_month
data_week = df_week
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objs as go
import plotly.express as px
```

```python
data.tail(10)
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x = data.index,
                        y = data['kasus_positif'],
                        mode = 'lines',
                        name = 'Kasus Positif',
                        marker_color = 'red'))


fig.add_trace(go.Scatter(x = data.index,
                        y = data['kasus_sembuh'],
                        mode = 'lines',
                        name = 'Kasus Sembuh',
                        marker_color = 'green'))


fig.add_trace(go.Scatter(x = data.index,
                        y = data['kasus_dirawat'],
                        mode = 'lines',
                        name = 'Kasus Dirawat',
                        marker_color = 'blue'))

fig.add_trace(go.Scatter(x = data.index,
                        y = data['kasus_meninggal'],
                        mode = 'lines',
                        name = 'Kasus Meninggal',
                        marker_color = 'brown'))

fig.update_layout(title = '<b> Data Harian Kasus Covid-19 di Indonesia <b>',
                  title_x = 0.5,
                  title_font = dict(size = 18, color = 'Darkred'),
                  template = 'plotly_white')

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x = data.index,
                        y = data['kumulatif_kasus_positif'],
                        mode = 'lines',
                        name = 'Kasus Positif',
                        marker_color = 'red'))


fig.add_trace(go.Scatter(x = data.index,
                        y = data['kumulatif_kasus_sembuh'],
                        mode = 'lines',
                        name = 'Kasus Sembuh',
                        marker_color = 'green'))


fig.add_trace(go.Scatter(x = data.index,
                        y = data['kumulatif_kasus_dirawat'],
                        mode = 'lines',
                        name = 'Kasus Dirawat',
                        marker_color = 'blue'))

fig.add_trace(go.Scatter(x = data.index,
                        y = data['kumulatif_kasus_meninggal'],
                        mode = 'lines',
                        name = 'Kasus Meninggal',
                        marker_color = 'brown'))

fig.update_xaxes(
    rangeslider_visible=False,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="todate"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)

fig.update_layout(title = '<b> Data Kumulatif Kasus Covid-19 di Indonesia <b>',
                  title_x = 0.5,
                  title_font = dict(size = 18, color = 'Darkred'),
                  template = 'plotly_white')

fig.show()
```

```python
fig1 = px.bar(data_week, x= data_week.index, y="kasus_positif", text='kasus_positif',
        title= 'Data Minguan Kasus Covid-19 di Indonesia')
fig1.update_traces(textposition='outside')
fig1.update_layout(uniformtext_minsize=8, uniformtext_mode='hide')

fig1.show()
```
