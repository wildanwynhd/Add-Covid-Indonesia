<h2 style='color:Yellow'> Get Data From API Postman  </h2>

```python
import requests
import pandas as pd
import numpy as np
resp = requests.get('https://data.covid19.go.id/public/api/prov.json')
print(resp)
print(resp.headers)
```

<h2 style='color:Yellow'> Check Length and Component of JSON </h2>

```python
dataRaw_ProvinceIndonesia = resp.json()
print('Length of dataRaw_ProvinceIndonesia : %d.' %len(dataRaw_ProvinceIndonesia))
print('Komponen dataRaw_ProvinceIndonesia  : %s.' %dataRaw_ProvinceIndonesia.keys())
dataRaw_ProvinceIndonesia
```

<h2 style='color:Yellow'> Access the List Data </h2>

```python
# Access the data in the JSON
listData_access = dataRaw_ProvinceIndonesia['list_data'] 
print('Length of CovidIndonesia_update : %d.' %len(listData_access))
listData_access
```

<h2 style='color:Yellow'> Change Format Data </h2>

```python
for listData_Data in listData_access:
    print(listData_Data)
```

<h2 style='color:Yellow'> Check Data Type </h2>

```python
print(type(listData_access))
print(type(listData_Data))
```

```python
# Data Usia
print('Example Data Kelompok Umur')
kelompokUmur_access = listData_Data['kelompok_umur']
for kelompokUmur_data in kelompokUmur_access:
    for usia_data in kelompokUmur_data['usia'].items():
        print(usia_data)
    
```

<h4 style='color:Yellow'>1. DataFrame Total Kasus yang terjadi di berbagai provinsi </h4>

```python
# Import 
import time
import pyodbc

# Create Timer
start_time = time.time()
from sqlalchemy import create_engine

# Data Frame
DataFrame1 = pd.DataFrame(listData_access, columns=['key', 'doc_count', 'jumlah_kasus', 'jumlah_sembuh', 'jumlah_meninggal', 'jumlah_dirawat'])
DataFrame1

# Connect database 
conn_str = (
    r'DRIVER={SQL Server Native Client 11.0};'
    r'SERVER=DESKTOP-CO0CEK4;'
    r'DATABASE=Covid19_Indonesia;'
    r'Trusted_Connection=yes;'
)

connect = pyodbc.connect(conn_str)

cursor = connect.cursor()

for index, row in DataFrame1.iterrows():
    cursor.execute('INSERT INTO dbo.Kasus_Covid([provinsi], [count], [totalKasus], [totalSembuh], [totalMeninggal], [totalDirawat] ) values (?,?,?,?,?,?)', 
                    row['key'], 
                    row['doc_count'],
                    row['jumlah_kasus'],
                    row['jumlah_sembuh'],
                    row['jumlah_meninggal'], 
                    row['jumlah_dirawat'])
    connect.commit()
cursor.close()
connect.close()

# see total time to do insert
print("%s seconds ---" % (time.time() - start_time))
```
